```python
BIT_WIDTH = 8
MANTISSA = 0
INPUTS_NUMBER = 5
SIGNED = True

FIRST_LAYER_NEURONS = 3
```

```python
# import numpy as np
# if SIGNED == True:
#     sequence = np.linspace(-((2**(4))/2),((2**4)/2) -1,int(2**4))
# else:
#     sequence = np.linspace(0,2**4 -1,int(2**4))
# sequence

import numpy as np
if SIGNED == True:
    sequence = np.linspace(-((2**BIT_WIDTH)/2),((2**BIT_WIDTH)/2) -1,int(2**BIT_WIDTH))
else:
    sequence = np.linspace(0,2**BIT_WIDTH -1,int(2**BIT_WIDTH))
sequence
```

```python
# criando valores base
df_inputs = pd.DataFrame()

for i in range(0,INPUTS_NUMBER*FIRST_LAYER_NEURONS):

    df_inputs[i] = sequence

df_inputs
```

```python
df_inputs_bin = df_inputs
from fxpmath import Fxp
for i in range (0 , INPUTS_NUMBER*FIRST_LAYER_NEURONS):

    df_inputs_bin[i] = (df_inputs[i].apply(lambda x:  Fxp(x, signed=True, n_word= BIT_WIDTH, n_frac= MANTISSA).bin() ) )
df_inputs_bin
```
